## Предобработка данных

### Cтиль заголовков

Рассмотрим названия столбцов в исходной таблице:

```python
df.columns
```

Приведем все заголовки к единому стилю:

```python
df.columns = df.columns.str.lower()
```

### Преобразование типов

Рассмотрим имеющиеся типы данных:

```python
df.dtypes
```

Заменим в столбце `user_score` значение `tbd` на `np.NaN`.
`tbd` - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено).

```python
df['user_score'].replace('tbd', np.NaN, inplace=True)
```

Заменим в некоторых столбцах тип данных с `float64` на `float16` для оптимизации памяти. Тип данных в столбце `year_of_release` заменим на `Int64`:

```python
for col in df.columns:
    if ('sales' in col) or ('score' in col):
        df[col] = df[col].astype('float16')
    if 'year' in col:
        df[col] = df[col].astype('Int64')

#df['year_of_release'] = df['year_of_release'].astype('Int64')
#df['critic_score'] = df['critic_score'].astype('float16')
#df['user_score'] = df['user_score'].astype('float16')
```

### Обработка пропусков

Рассмотрим количество пропусков в каждом столбце:

```python
df.isna().sum()
```

**Наибольшее** количество пропусков обнаружено в столбцах, описывающих рейтинг критиков и игроков, а также в столбце, описывающем возрастные ограничения. 

**Наименьшее** количество пропусков обнаружено в столбцах с именем и жанром - всего лишь 2 пропуска.

Рассмотрим пропуски в столбцах `name` и `genre`:

```python
df[df['name'].isna()]
```

Заметим, что пропущенные данные в этих столбцах парные. Избавимся от данных пропусков, так как процент от количества данных составляяет **менее 1%**.

```python
df.dropna(subset=['name', 'genre'], inplace=True)
```

Рассмотрим пропуски в столбце `year_of_release`:

```python
df[df['year_of_release'].isna()]
```

Многие игры были выпущены для нескольких платформ. Заменим пропущенные значения года выпуска, на год выпуска такой же игры, но у другой платформы:

```python
df_grouped = df.groupby(['platform', 'name'])['year_of_release'].median()
df_grouped = df_grouped.reset_index()
```

```python
def fill_year(row):
    year = row['year_of_release']
    name = row['name']
    if pd.isna(year):
        return df_grouped[df_grouped['name'] == name]['year_of_release'].min()
    return year
```

```python
df['year_of_release'].fillna(df.apply(fill_year, axis=1), inplace=True)
```

Проверим, сколько пропусков удалось заменить:

```python
df.isna().sum()
```

Удалось обработать более ста пропусков. Пропущенные значения в остальных стобцах оставим нетронутыми, чтобы не исказить данные.

### Расчет суммарных продаж

Создадим столбец `total_sales`, в котором рассчитаем продажи игры по всем регионам:

```python
df['total_sales'] = df[['na_sales','eu_sales','jp_sales','other_sales']].sum(axis=1)
```

Изучим взаимосвязь между глобальными продажами и оценками как пользователей, так и критиков. Для этого создадим функцию, вычисляющую необходимые коэффициенты корреляции и строящую диаграммы рассеяния:

```python
def corr_analys(platform):
    data = df_actual.query("platform == @platform")
    total_sales = data['total_sales']
    user_score = data['user_score']
    critic_score = data['critic_score']
    
    corr_ts_us = total_sales.corr(user_score)
    corr_ts_cs = total_sales.corr(critic_score)
    
    print(f'Коэффициент корреляции между глобальными продажами и оценкой игроков: {corr_ts_us}')
    print(f'Коэффициент корреляции между глобальными продажами и оценкой критиков: {corr_ts_cs}')
    
    sns.scatterplot(x=total_sales, y=user_score, color='r').set(title='Диграмма рассеяения',
                                                       xlabel='Глобальные продажи, млн.копий',
                                                       ylabel='Оценка пользователей',                        
                                                    )
    plt.show()
    sns.scatterplot(x=total_sales, y=critic_score).set(title='Диграмма рассеяения',
                                                       xlabel='Глобальные продажи, млн.копий',
                                                       ylabel='Оценка критиков'
                                                    )
    plt.show()
```

```python
for platform in df_actual['platform'].unique():
    print(f'Анализ для платформы {platform}:')
    corr_analys(platform)
```

Анализ корреляции показал, что для каждой из перечисленных платформ, корреляция между продажами и оценками как пользователей, так и критиков очень мала. Тем не менее, можно сказать, что оценка критиков является более весомым аргументом для покупки, чем оценка пользователей.

Однако среди всех результатов выбивается платформа **WiiU**, у которой коэффициент корреляции между глобальными продажами и оценкой игроков выше, нежели корреляцией между оценкой продаж и отзывов критиков.

Также анализ показал неккоректность результатов для платформ **PSP**, **WII** и **DS**, что возможно обусловлено отсутствием выпущенных игр за актуальный период, их критически низкими продажами или же отсутствием оценок:

## Портрет пользователя

Составим портрет пользователя каждого региона.

Для последующего изучения влияния возрастного рейтинга на продажи создадим функцию `age_restr`, которая расшифровывает метки `ESRB` для определения допустимого возраста:

```python
def age_restr(rating):
    if rating == 'M':
        return '17+'
    elif rating == 'T':
        return '13+'
    elif rating == 'E':
        return '6+'
    elif rating == 'E10':
        return '10+'
    else:
        return 'Нет рейтинга'
```

Создадим столбец `allowed_age` с допустимым возрастом для игры:

```python
df_actual['allowed_age'] = df_actual['rating'].apply(age_restr)
```

Создадим функцию, которая позволит получить необходимые данные и распределения продаж по платформам, жанрам и возрастному рейтингу:

```python
def profiler(region, distr):
    if region == 'EU':
        platform_sales_reg = df_actual.pivot_table(index='platform', values='eu_sales', aggfunc='sum').sort_values(
        by='eu_sales', ascending=False)
        
        genre_sales_reg = df_actual.pivot_table(index='genre', values='eu_sales', aggfunc='median').sort_values(
        by='eu_sales', ascending=False)     
        
        age_rate_sales = df_actual.pivot_table(index='allowed_age', values='eu_sales', aggfunc='sum').sort_values(
        by='eu_sales', ascending=False) 
        
    elif region == 'NA':
        platform_sales_reg = df_actual.pivot_table(index='platform', values='na_sales', aggfunc='sum').sort_values(
        by='na_sales', ascending=False)
        
        genre_sales_reg = df_actual.pivot_table(index='genre', values='na_sales', aggfunc='median').sort_values(
        by='na_sales', ascending=False)
        
        age_rate_sales = df_actual.pivot_table(index='allowed_age', values='na_sales', aggfunc='sum').sort_values(
        by='na_sales', ascending=False) 
        
    elif region == 'JP':
        platform_sales_reg = df_actual.pivot_table(index='platform', values='jp_sales', aggfunc='sum').sort_values(
        by='jp_sales', ascending=False)
        
        genre_sales_reg = df_actual.pivot_table(index='genre', values='jp_sales', aggfunc='median').sort_values(
        by='jp_sales', ascending=False)
        
        age_rate_sales = df_actual.pivot_table(index='allowed_age', values='jp_sales', aggfunc='sum').sort_values(
        by='jp_sales', ascending=False) 
    
    platform_sales_reg.columns = ['total_sales']
    genre_sales_reg.columns = ['total_sales']
    age_rate_sales.columns = ['total_sales']
    
    if distr == 'platform':
        return platform_sales_reg
    elif distr == 'genre':
        return genre_sales_reg
    else:
        return age_rate_sales
```

## Проверка статистических гипотез

Создадим функцию `check_hypot`, принимающий два набора данных, уровень статистической значимости `a` и тип гипотезы:

```python
def check_hypot(data1, data2, a, hypot):
    sns.distplot(data1, bins=25)
    sns.distplot(data2, bins=25)
    plt.xlabel('Оценка пользователей')
    plt.ylabel('Плотность вероятности')
    
    if hypot == 'platform':
        plt.title('Распределение оценки пользователей в зависимости от платформы')
    elif hypot == 'genre':
        plt.title('Распределение оценки пользователей в зависимости от жанра')
    
    print(f'Дисперсия 1-й генеральной совокупности: {np.var(data1)}')
    print(f'Дисперсия 2-й генеральной совокупности: {np.var(data2)}')
    
    print(f'Среднее 1-й генеральной совокупности: {np.mean(data1)}')
    print(f'Среднее 2-й генеральной совокупности: {np.mean(data2)}')
    
    
    results = st.ttest_ind(data1, data2, equal_var=False)
    
    print(f'p-value: {results.pvalue}')
    
    if (results.pvalue < a):
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не удалось отвернуть нулевую гипотезу')
        
        
```

### Гипотеза о средних пользовательских оценках платформ `Xbox One` и `PC`

$H_{0}$ - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

$H_{1}$ - Средние пользовательские рейтинги платформ Xbox One и PC не одинаковые;

$\alpha$ = 0.05

Определим необходимые наборы данных:

```python
xone_users = df_actual.query("platform == 'XOne'")['user_score'].dropna()
pc_users = df_actual.query("platform == 'PC'")['user_score'].dropna()
```

Проверим гипотезу:

```python
check_hypot(xone_users, pc_users, a=0.05, hypot='platform')
```

На графике **синим** цветом обозначено распределение платформы `Xbox One`, **красным** - жанра `PC`.

### Гипотеза о средних пользовательских оценках жанров `Action` и `Sports`

$H_{0}$ - Средние пользовательские рейтинги жанров `Action` и `Sports` одинаковые;

$H_{1}$ - Средние пользовательские рейтинги жанров `Action` и `Sports` не одинаковые;

$\alpha$ = 0.05

Определим необходимые наборы данных:

```python
action_games = df_actual.query("genre == 'Action'")['user_score'].dropna()
sport_games = df_actual.query("genre == 'Sports'")['user_score'].dropna()
```

Проверим гипотезу:

```python
check_hypot(action_games, sport_games, a=0.05, hypot='platform')
```

На графике **синим** цветом обозначено распределение жанра `Action`, **красным** - жанра `Sports`.

### Вывод

При проверке статистических гипотез удалость выяснить следующие факты:

* Средние пользовательские оценки платформ `Xbox One` и `PC` не различаются. Для пользователей платформы `Xbox One` средняя оценка составила **6.52**, а для пользователей `PC` - **6.28**
